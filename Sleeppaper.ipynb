```python
#preparing the data for analysis 

import pandas as pd 
newdf = pd.read_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/newdf.csv')
newerdf=newdf.copy()
print(len(newerdf))

newerdf.pop('gender')
newerdf.rename(columns={'gender-quantised':'gender'}, inplace=True)
newerdf.rename(columns={'naps-quantised':'frequency_of_naps'}, inplace=True)
newerdf.rename(columns={'activity-quantised':'hours_exercise_per_week_in_last_6_months'}, inplace=True)
newerdf.rename(columns={'education-quantised':'highest_education_level_acheived'}, inplace=True)
newerdf.rename(columns={'sleep':'hours_of_sleep'}, inplace=True)
newerdf.rename(columns={'wakeup_times':'number_of_times_wakeup_during_night'}, inplace=True)
newerdf.rename(columns={'time_awake':'time_spent_awake_during_night_mins'}, inplace=True)
newerdf.rename(columns={'nap_time':'nap_duration_mins'}, inplace=True)
newerdf.rename(columns={'alarm-quantised':'frequency_of_alarm_usage'}, inplace=True)
newerdf.rename(columns={'alcohol':'drink_alcohol_yes/no'}, inplace=True)
newerdf.rename(columns={'caffeine':'drink_caffeine_yes/no'}, inplace=True)
newerdf.rename(columns={'exercise ':'hours_exercise_per_week'}, inplace=True)
newerdf.rename(columns={'phone_hours':'hours_of_phone_use_per_week'}, inplace=True)
newerdf.rename(columns={'video_game_phone_hours':'video_game_phone_tablet_hours_per_week'}, inplace=True)
newerdf.rename(columns={'video_game_hours':'video_game_all_devices_hours_per_week'}, inplace=True)
newerdf.rename(columns={'drink_caffeine_yes/no':'drink_caffeine_yes_no'}, inplace=True)
newerdf.rename(columns={'drink_alcohol_yes/no':'drink_alcohol_yes_no'}, inplace=True)

import pandas as pd 

#calculate the time of going to sleep (hour), the time to fall asleep (mins) and the normal time of waking up (hour)

newerdf['time_of_going_to_sleep'] = newerdf['time-hour'] + newerdf['time-minute']/60
newerdf['normal_time_of_wakeup'] = newerdf['time_wakeup-hour'] + newerdf['time_wakeup-minute']/60

#calculate time in bed in hours and account for minus values by adding 24

for i in range(0,len(newerdf['time_of_going_to_sleep'])):
    if (newerdf['time_of_going_to_sleep'].iloc[i] > newerdf['normal_time_of_wakeup'].iloc[i]):
      if newerdf['time_of_going_to_sleep'].iloc[i] <= 12:
         newerdf['normal_time_of_wakeup'].iloc[i] = newerdf['normal_time_of_wakeup'].iloc[i]+12
      if newerdf['time_of_going_to_sleep'].iloc[i] > 12:
         newerdf['normal_time_of_wakeup'].iloc[i] = newerdf['normal_time_of_wakeup'].iloc[i]+24

newerdf['time_in_bed_hours'] = newerdf['normal_time_of_wakeup'] - newerdf['time_of_going_to_sleep']

#calculate time in bed in mims 

newerdf['time_in_bed_mins'] = newerdf['time_in_bed_hours']*60

#turn naps from categorical variable (frequency) to binary (yes/no)

newerdf['takes_naps_yes/no'] = newerdf['nap_duration_mins']
newerdf['takes_naps_yes/no'][newerdf['nap_duration_mins']==0] = 0 
newerdf['takes_naps_yes/no'][~(newerdf['nap_duration_mins']==0)] = 1 
newerdf.rename(columns={'takes_naps_yes/no':'takes_naps_yes_no'}, inplace=True)

#turn alarm usage from categorical variable (frequency) to binary (yes/no)

newerdf['use_alarms_yes_no'] = newerdf['alarm']
newerdf['use_alarms_yes_no'] = pd.Categorical(newerdf['use_alarms_yes_no'])
newerdf['use_alarms_yes_no'] = newerdf['use_alarms_yes_no'].cat.codes
newerdf['use_alarms_yes_no'][newerdf['use_alarms_yes_no']==4] = 5
newerdf['use_alarms_yes_no'][~(newerdf['use_alarms_yes_no']==5)] = 6

#calculate the sleepiness resolution index, adding '1' to the values for 'sleepiness on bed' and 'sleepiness waking' to ensure one is not dividing by zero

newerdf['sleepiness_bed'] = newerdf['sleepiness_bed']+1
newerdf['sleepiness_waking'] = newerdf['sleepiness_waking']+1
newerdf['sleepiness_resolution_index'] = newerdf['sleepiness_bed']/newerdf['sleepiness_waking']

#removing those with hours that are less than 4 hours or hours of sleep that exceed 12 hours

correcteddf = newerdf.copy()

#remove those who selected 'other' for gender 

print(len(correcteddf[correcteddf['gender']==3.0]))
correcteddf = correcteddf[~(correcteddf['gender']==3.0)]

#create a normalised TIB score 

correcteddf['normalised_TIB'] = ((correcteddf['time_spent_awake_during_night_mins']/60)/correcteddf['time_in_bed_hours'])*100

#combine height in feet and inches to ensure it is one measurement and combine sunlight minute and hour into one value 

correcteddf['height_feet_amend'] = correcteddf['height_feet']*12
correcteddf['height'] = correcteddf['height_inches'] + correcteddf['height_feet_amend']
correcteddf['sunlight'] = correcteddf['sunlight-hour'] + correcteddf['sunlight-minute']/60

#calculating BMI from height and weight 

correcteddf['heightmetres'] = correcteddf['height_feet']*0.3048 + correcteddf['height_inches']*0.0254
correcteddf['weight_kilo'] = correcteddf['weight']*0.453592
correcteddf['BMI'] = correcteddf['weight_kilo'] / correcteddf['heightmetres']**2

#convert values from 0 and above (represent morning time/am) to values exceeding 24 to ensure that the data follow the appropriate distribution

indextimesleep = correcteddf.index[(correcteddf['time_of_going_to_sleep']>=0)&(correcteddf['time_of_going_to_sleep']<12)]
indextimewakeup = correcteddf.index[(correcteddf['normal_time_of_wakeup']>=0)&(correcteddf['normal_time_of_wakeup']<12)]  
correcteddf.loc[indextimesleep.values,'time_of_going_to_sleep'] = correcteddf.loc[indextimesleep.values,'time_of_going_to_sleep']+24
correcteddf.loc[indextimewakeup.values,'normal_time_of_wakeup'] = correcteddf.loc[indextimewakeup.values,'normal_time_of_wakeup']+24

#binarising 'use_alarms_yes_no' to either 0 or 1  

correcteddf['use_alarms_yes_no'][correcteddf['use_alarms_yes_no'] == 6] = 1
correcteddf['use_alarms_yes_no'][correcteddf['use_alarms_yes_no'] == 5] = 0

#converting caffeine/alcohol to a binary variable 

correcteddf['drink_alcohol_yes_noscale'] = correcteddf['drink_alcohol_yes_no']
correcteddf['drink_caffeine_yes_noscale'] = correcteddf['drink_caffeine_yes_no']
correcteddf['drink_caffeine_yes_no'][correcteddf['drink_caffeine_yes_no'] == 0] = 0
correcteddf['drink_caffeine_yes_no'][~(correcteddf['drink_caffeine_yes_no'] == 0)] = 1
correcteddf['drink_alcohol_yes_no'][correcteddf['drink_alcohol_yes_no'] == 0] = 0
correcteddf['drink_alcohol_yes_no'][~(correcteddf['drink_alcohol_yes_no'] == 0)] = 1

#reverse coding the sleep quality 

correcteddf['sleep_quality'] = 11-correcteddf['sleep_quality']
correcteddf['sleep_fragmentation'] = None

#calculating sleep fragmentation 

for i in range(0,len(correcteddf['time_in_bed_hours'])): 
   if correcteddf['time_in_bed_hours'].iloc[i] >= correcteddf['time_spent_awake_during_night_mins'].iloc[i]:
      correcteddf['sleep_fragmentation'].iloc[i] = correcteddf['time_spent_awake_during_night_mins'].iloc[i]/correcteddf['time_in_bed_hours'].iloc[i]
   else:
      correcteddf['sleep_fragmentation'].iloc[i] = 'Remove this value'

#changing name of 'fall_asleep' column

correcteddf = correcteddf.rename(columns={'fall_asleep':'difficultywakeup'})

#converting correcteddf to a csv file 

correcteddf.to_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/correcteddf'+'.csv')
```

```python
#preprocessing data for 6-9 hours sleep 

#preparing the data for analysis 

import pandas as pd 
newdf = pd.read_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/newdf.csv')
newerdf=newdf.copy()

newerdf.pop('gender')
newerdf.rename(columns={'gender-quantised':'gender'}, inplace=True)
newerdf.rename(columns={'naps-quantised':'frequency_of_naps'}, inplace=True)
newerdf.rename(columns={'activity-quantised':'hours_exercise_per_week_in_last_6_months'}, inplace=True)
newerdf.rename(columns={'education-quantised':'highest_education_level_acheived'}, inplace=True)
newerdf.rename(columns={'sleep':'hours_of_sleep'}, inplace=True)
newerdf.rename(columns={'wakeup_times':'number_of_times_wakeup_during_night'}, inplace=True)
newerdf.rename(columns={'time_awake':'time_spent_awake_during_night_mins'}, inplace=True)
newerdf.rename(columns={'nap_time':'nap_duration_mins'}, inplace=True)
newerdf.rename(columns={'alarm-quantised':'frequency_of_alarm_usage'}, inplace=True)
newerdf.rename(columns={'alcohol':'drink_alcohol_yes/no'}, inplace=True)
newerdf.rename(columns={'caffeine':'drink_caffeine_yes/no'}, inplace=True)
newerdf.rename(columns={'exercise ':'hours_exercise_per_week'}, inplace=True)
newerdf.rename(columns={'phone_hours':'hours_of_phone_use_per_week'}, inplace=True)
newerdf.rename(columns={'video_game_phone_hours':'video_game_phone_tablet_hours_per_week'}, inplace=True)
newerdf.rename(columns={'video_game_hours':'video_game_all_devices_hours_per_week'}, inplace=True)
newerdf.rename(columns={'drink_caffeine_yes/no':'drink_caffeine_yes_no'}, inplace=True)
newerdf.rename(columns={'drink_alcohol_yes/no':'drink_alcohol_yes_no'}, inplace=True)

import pandas as pd 

#calculate the time of going to sleep (hour), the time to fall asleep (mins) and the normal time of waking up (hour)

newerdf['time_of_going_to_sleep'] = newerdf['time-hour'] + newerdf['time-minute']/60
newerdf['normal_time_of_wakeup'] = newerdf['time_wakeup-hour'] + newerdf['time_wakeup-minute']/60

#calculate time in bed in hours and account for minus values by adding 24

for i in range(0,len(newerdf['time_of_going_to_sleep'])):
    if (newerdf['time_of_going_to_sleep'].iloc[i] > newerdf['normal_time_of_wakeup'].iloc[i]):
      if newerdf['time_of_going_to_sleep'].iloc[i] <= 12:
         newerdf['normal_time_of_wakeup'].iloc[i] = newerdf['normal_time_of_wakeup'].iloc[i]+12
      if newerdf['time_of_going_to_sleep'].iloc[i] > 12:
         newerdf['normal_time_of_wakeup'].iloc[i] = newerdf['normal_time_of_wakeup'].iloc[i]+24

newerdf['time_in_bed_hours'] = newerdf['normal_time_of_wakeup'] - newerdf['time_of_going_to_sleep']

#calculate time in bed in mims 

newerdf['time_in_bed_mins'] = newerdf['time_in_bed_hours']*60

#turn naps from categorical variable (frequency) to binary (yes/no)

newerdf['takes_naps_yes/no'] = newerdf['nap_duration_mins']
newerdf['takes_naps_yes/no'][newerdf['nap_duration_mins']==0] = 0 
newerdf['takes_naps_yes/no'][~(newerdf['nap_duration_mins']==0)] = 1 
newerdf.rename(columns={'takes_naps_yes/no':'takes_naps_yes_no'}, inplace=True)

#turn alarm usage from categorical variable (frequency) to binary (yes/no)

newerdf['use_alarms_yes_no'] = newerdf['alarm']
newerdf['use_alarms_yes_no'] = pd.Categorical(newerdf['use_alarms_yes_no'])
newerdf['use_alarms_yes_no'] = newerdf['use_alarms_yes_no'].cat.codes
newerdf['use_alarms_yes_no'][newerdf['use_alarms_yes_no']==4] = 5
newerdf['use_alarms_yes_no'][~(newerdf['use_alarms_yes_no']==5)] = 6

#calculate the sleepiness resolution index, adding '1' to the values for 'sleepiness on bed' and 'sleepiness waking' to ensure one is not dividing by zero

newerdf['sleepiness_bed'] = newerdf['sleepiness_bed']+1
newerdf['sleepiness_waking'] = newerdf['sleepiness_waking']+1
newerdf['sleepiness_resolution_index'] = newerdf['sleepiness_bed']/newerdf['sleepiness_waking']

#removing those with hours that are less than 4 hours or hours of sleep that exceed 12 hours

correcteddf = newerdf.copy()

#create a normalised TIB score 

correcteddf['normalised_TIB'] = ((correcteddf['time_spent_awake_during_night_mins']/60)/correcteddf['time_in_bed_hours'])*100

#combine height in feet and inches to ensure it is one measurement and combine sunlight minute and hour into one value 

correcteddf['height_feet_amend'] = correcteddf['height_feet']*12
correcteddf['height'] = correcteddf['height_inches'] + correcteddf['height_feet_amend']
correcteddf['sunlight'] = correcteddf['sunlight-hour'] + correcteddf['sunlight-minute']/60

#calculating BMI from height and weight 

correcteddf['heightmetres'] = correcteddf['height_feet']*0.3048 + correcteddf['height_inches']*0.0254
correcteddf['weight_kilo'] = correcteddf['weight']*0.453592
correcteddf['BMI'] = correcteddf['weight_kilo'] / correcteddf['heightmetres']**2

#convert values from 0 and above (represent morning time/am) to values exceeding 24 to ensure that the data follow the appropriate distribution

indextimesleep = correcteddf.index[(correcteddf['time_of_going_to_sleep']>=0)&(correcteddf['time_of_going_to_sleep']<12)]
indextimewakeup = correcteddf.index[(correcteddf['normal_time_of_wakeup']>=0)&(correcteddf['normal_time_of_wakeup']<12)]  
correcteddf.loc[indextimesleep.values,'time_of_going_to_sleep'] = correcteddf.loc[indextimesleep.values,'time_of_going_to_sleep']+24
correcteddf.loc[indextimewakeup.values,'normal_time_of_wakeup'] = correcteddf.loc[indextimewakeup.values,'normal_time_of_wakeup']+24

#binarising 'use_alarms_yes_no' to either 0 or 1  

correcteddf['use_alarms_yes_no'][correcteddf['use_alarms_yes_no'] == 6] = 1
correcteddf['use_alarms_yes_no'][correcteddf['use_alarms_yes_no'] == 5] = 0

#converting caffeine/alcohol to a binary variable 

correcteddf['drink_alcohol_yes_noscale'] = correcteddf['drink_alcohol_yes_no']
correcteddf['drink_caffeine_yes_noscale'] = correcteddf['drink_caffeine_yes_no']
correcteddf['drink_caffeine_yes_no'][correcteddf['drink_caffeine_yes_no'] == 0] = 0
correcteddf['drink_caffeine_yes_no'][~(correcteddf['drink_caffeine_yes_no'] == 0)] = 1
correcteddf['drink_alcohol_yes_no'][correcteddf['drink_alcohol_yes_no'] == 0] = 0
correcteddf['drink_alcohol_yes_no'][~(correcteddf['drink_alcohol_yes_no'] == 0)] = 1

#reverse coding the sleep quality 

correcteddf['sleep_quality'] = 11-correcteddf['sleep_quality']
correcteddf['sleep_fragmentation'] = None

#calculating sleep fragmentation 

for i in range(0,len(correcteddf['time_in_bed_hours'])): 
   if correcteddf['time_in_bed_hours'].iloc[i] >= correcteddf['time_spent_awake_during_night_mins'].iloc[i]:
      correcteddf['sleep_fragmentation'].iloc[i] = correcteddf['time_spent_awake_during_night_mins'].iloc[i]/correcteddf['time_in_bed_hours'].iloc[i]
   else:
      correcteddf['sleep_fragmentation'].iloc[i] = 'Remove this value'

#changing name of 'fall_asleep' column

correcteddf = correcteddf.rename(columns={'fall_asleep':'difficultywakeup'})

#correcteddf hours of sleep restriction 

correcteddf = correcteddf[(correcteddf['hours_of_sleep'] >= 6.0) & (correcteddf['hours_of_sleep'] <= 9.0)]

#converting correcteddf to a csv file 

correcteddf.to_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/correcteddfRESTRICTED'+'.csv')
```

```python
#preparing the correcteddf for analysis RESTRICTED

correcteddf = pd.read_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/correcteddfRESTRICTED'+'.csv')

import pandas as pd
import numpy as np 
import seaborn as sns 

correcteddf = correcteddf.rename(columns={'fall_asleep':'difficultywakeup'})
correcteddf = correcteddf[~(correcteddf['gender']==3.0)]

#calculating Mahalanobis Distance

from scipy.stats import chi2
  
def calculateMahalanobis(y=None, data=None, cov=None):
  
    y_mu = y - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(y_mu, inv_covmat)
    mahal = np.dot(left, y_mu.T)
    return mahal.diagonal()
    
correcteddf['Mahalanobis'] = calculateMahalanobis(y=correcteddf[['age','hours_of_phone_use_per_week','frequency_of_naps','frequency_of_alarm_usage','hours_of_sleep','number_of_times_wakeup_during_night','time_spent_awake_during_night_mins','nap_duration_mins','sleep_quality','difficultywakeup','sleepiness_waking','sleepiness_bed','time_in_bed_hours','sleepiness_resolution_index','video_game_all_devices_hours_per_week','drink_caffeine_yes_noscale','drink_alcohol_yes_noscale','BMI','sunlight']],data=correcteddf[['age','hours_of_phone_use_per_week','frequency_of_naps','frequency_of_alarm_usage','hours_of_sleep','number_of_times_wakeup_during_night','time_spent_awake_during_night_mins','nap_duration_mins','sleep_quality','difficultywakeup','sleepiness_waking','sleepiness_bed','time_in_bed_hours','sleepiness_resolution_index','video_game_all_devices_hours_per_week','drink_caffeine_yes_noscale','drink_alcohol_yes_noscale','BMI','sunlight']])
correcteddf['m_p'] = 1 - chi2.cdf(correcteddf['Mahalanobis'], 19)

#calculate line plot for mahalanobis distance

sns.lineplot(x=chi2.cdf(correcteddf['Mahalanobis'], 19),y=correcteddf['Mahalanobis'])

#storing the Mahalanobis values in a dataframe and calculating the cut-off value 

malaxvariable=chi2.cdf(correcteddf['Mahalanobis'], 19)
malayvariable=correcteddf['Mahalanobis']
malayvariable=pd.DataFrame(malayvariable)
malayvariable = malayvariable.reset_index(drop=True)
malayvariable.columns=['malayvals']
malaxvariable=pd.DataFrame(malaxvariable)
malaxvariable.columns=['malaxvals']
malaxvariable = malaxvariable.reset_index(drop=True)
mal = pd.concat([malayvariable,malaxvariable],axis=1)
mal.sort_values(by='malaxvals',ascending=True)
mal = mal.round(1)
minval = mal[mal['malaxvals']==1.0].min()[0]

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt 
import seaborn as sns

correcteddf = correcteddf[correcteddf['Mahalanobis']<=minval]
correcteddf = correcteddf.rename(columns={"response-1 -quantised":'response1q',"response-2-quantised":'response2q',"response-3-quantised":'response3q',"response-4-quantised":'response4q',"response-5-quantised":'response5q',"response-6-quantised":'response6q',"response-7-quantised":'response7q',"response-8-quantised":'response8q',"response-9-quantised":'response9q',"response-10-quantised":'response10q',"response-11-quantised":'response11q',"response-12-quantised":'response12q',"response-13-quantised":'response13q',"response-14-quantised":'response14q',"gps1-quantised":'gps1q',"gps2-quantised":'gps2q',"gps3-quantised":'gps3q',"gps4-quantised":'gps4q',"gps6-quantised":'gps6q',"gps7-quantised":'gps7q'})
correcteddf.to_csv(r'/Users/macbook/Desktop/UCL PhD Work/SHQoutputs/PREPROCESSING/correcteddfRESTRICTED.csv')
```
